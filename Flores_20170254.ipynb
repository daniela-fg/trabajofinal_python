## Transformación de datos

 


### Reducción de base de datos
Con el fin de permitir que el programa se ejecute más rápido, se reducirán las variables a usar. Se creará una base de datos más pequeña.

Para el archivo 'Cultura', se seleccionarán las variables ```CONGLOME, VIVIENDA, HOGAR, DOMINIO, P606 y P606N```. 



````python
#1. Lista con columnas a usar
lst1 = ['CONGLOME', 'VIVIENDA', 'HOGAR', 'DOMINIO', 'P606', 'P606N'] #para Cultura



#2. Creando subset
BD = Cultura[lst1].copy().reset_index()
````

````python
BD
````

````python
BD['P606'].value_counts()
````

### Eliminación de valores perdidos/vacíos

Luego, a fin de poder realizar los cruces de información propuestos, se eliminarán las filas con datos vacíos o perdidos.

````python
#Eliminación de observaciones que no tienen información sobre bienes adquiridos (P606). Valores 2 y 9. 
BD = BD[BD.P606 != 2]
BD = BD[BD.P606 != 9]
````

````python
BD
````

### Añadido de etiquetas

````python
#1- Se agregarán etiquetas correspondientes a la variable Gran Dominio
BD['DOMINIO2'] = BD['DOMINIO']

BD.replace({'DOMINIO2': {1: 'Costa Norte', 2: 'Costa Centro', 3:'Costa Sur', 4:'Sierra Norte',
                         5:'Sierra Centro', 6:'Sierra Sur', 7:'Selva', 8:'Lima Metropolitana'}}, inplace = True)

BD
````

````python
#2- Se agregarán etiquetas correspondientes a la variable P606N
BD['P606CAT'] = BD['P606N']

BD.replace({'P606CAT': {1: 'Periodicos, revistas y libros', 
                        2: 'Esparcimiento y diversion (Cine, CD, juguetes, espectaculos deportivos, etc) ', 
                        3:'Aportaciones o cuotas a clubes, asociaciones, etc.', 
                        4:'Gastos en gimnasio (Areobicos, artes marciales, pesas)',
                         5:'USB', 6:'Articulos deportivos', 
                        7:'Utiles de escritorio (Cuadernos, papeles, etc.)', 8:'Otros gastos'}}, inplace = True)

BD
````

### Haciendo subset como último paso previo al gráfico

````python
#Como se mencionó en la introducción, deseamos averiguar en qué región geográfica del país se consumen más Periódicos, revistas y libros. Para ello, crearemos un subset. 
BD2 = BD[BD['P606CAT'] == "Periodicos, revistas y libros"]
````

````python
#Luego, se hace la selección para Periodicos, revistas y libros
BD3 = BD2[['P606N', 'DOMINIO2']].groupby('DOMINIO2').sum()
BD3
````

````python
#Finalmente, se crea un data frame en base a estas observaciones:
BD4 = {"Dominio Geográfico": ["Costa Centro", "Costa Norte", "Costa Sur", "Lima Metropolitana", "Selva", "Sierra Centro", "Sierra Norte", "Sierra Sur"], 
     "Cantidad": [1218, 915, 402, 1240, 942, 727, 113, 707]}

BD5 = pd.DataFrame(BD4)
````

````python
#Luego convertimos a porcentaje para hacer el gráfico más interesante

BD5['Porcent'] = (BD5['Cantidad'] / 
                  BD5['Cantidad'].sum()) * 100

BD5['Porcentaje'] = (19.4, 14.61, 6.42, 19.80, 15.04, 11.60, 1.80, 11.29)

BD5
````

## Gráfico

A diferencia de indicadores numéricos o tablas, un gráfico permitirá analizar el consumo y uso de productos o servicios según región de una manera más clara y atractiva. Es por ello que, a continuación, se procederá a construir uno en base a las variables ```Porcentaje``` y ```Dominio Geográfico```, que como se vio anteriormente, son el resultado de las recodificaciones de las variables ```P606``` y ```DOMINIO``` y la creación de subsets. Este paso previo facilitará el armado del gráfico.

Por otro lado, en tanto se está trabajando con variables cualitativas, se ha considerado que el mejor método para visualizar los datos será mediante un gráfico de pie que muestre los porcentajes. Por ende, este gráfico tiene leyenda y etiquetas, mas no ejes. 

````python

base = alt.Chart(BD5).encode(
    theta=alt.Theta("Porcentaje:Q", stack=True), color=alt.Color("Dominio Geográfico:N")
).properties(title={
      "text": ["Consumo de Periódicos, revistas y libros según Dominio Geográfico (Porcentaje)"], 
      "subtitle": ["Fuente: ENAHO 2020"],
      "color": "Black",
      "subtitleColor": "Black"
    })

pie = base.mark_arc(outerRadius=120)
text = base.mark_text(radius=140, size=20).encode(text="Porcentaje:N")

pie2 = pie + text
````
